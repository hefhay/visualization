```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading of dataset
data = pd.read_excel(r"C:\Users\Ahmed\Documents\Data Visualization dataset.xlsx")
```

```python
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
```

```python
data.head()
```

```python
# Group by year and sum sales
sales_by_year = data.groupby('Year')['Sales'].sum().reset_index()
sales_by_year
```

```python
# Plotting  a line graph for sales_by _year
plt.figure(figsize=(10, 6))
plt.plot(sales_by_year['Year'], sales_by_year['Sales'], marker='o', color='b', linestyle='-')

# Adding title and labels
plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')

# Set x-axis ticks as integers
plt.xticks(sales_by_year['Year'])

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by year and sum sales
sales_by_segment = data.groupby('Segment')['Sales'].sum().reset_index()
sales_by_segment 
```

```python
# Plotting
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(sales_by_segment['Sales'], labels=sales_by_segment['Segment'], autopct='%1.1f%%', startangle=90)

# Adding title
plt.title('Sales by Segment')

# Display the plot
 plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Group by category and sum sales
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
# Plotting a bar graph
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(sales_by_category['Category'], sales_by_category['Sales'], color='blue')

# Adding title and labels
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group by sub-category and sum sales
sales_by_sub_category = data.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_sub_category
```

```python
# Plotting
plt.figure(figsize=(12, 8))  # Set the figure size
plt.bar(sales_by_sub_category['Sub-Category'], sales_by_sub_category['Sales'], color='green')

# Adding title and labels
plt.title('Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Plotting the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data['Quantity'])

# Adding title and labels
plt.title('Box Plot of Sales')
plt.ylabel('Sales')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' DataFrame contains the required data
# data = ...

# Group by year and sum sales and quantity
sales_quantity_by_year = data.groupby('Year')[['Sales', 'Quantity']].sum().reset_index()

# Plotting the bar graph
plt.figure(figsize=(12, 6))

# Plotting sales
plt.bar(sales_quantity_by_year['Year'] - 0.2, sales_quantity_by_year['Sales'], width=0.4, label='Sales', color='skyblue')

# Plotting quantity
plt.bar(sales_quantity_by_year['Year'] + 0.2, sales_quantity_by_year['Quantity'], width=0.4, label='Quantity', color='orange')

# Adding title and labels
plt.title('Sales and Quantity by Year')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.xticks(sales_quantity_by_year['Year'])
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
